```python
#DAY 2

student_name = "Indhira"
study_hours = 4.5
passed_exam = True

print(f"Student name: {student_name}")
print(f"Studied for: {study_hours} hours")
print(f"Passed the exam: {passed_exam}")

############################################## 

hours_list = [2.5, 5.1, 3.2, 8.5, 3.5]
scores_list = [21, 47, 27, 75, 30]

high_study_count = 0

for i in hours_list:
    if (i > 4.0):
        high_study_count += 1
print(high_study_count)

############################################## 

def estimate_score(hours):
    score = 10 + (hours * 8)
    if score > 100:
        score = 100
    return score

result = estimate_score(6.0)
print(result)

############################################## 

def check_passing(score):
    if score >= 40:
        return "Pass"
    else:
        return "Fail"

for score in scores_list:
    result = check_passing(score)
    print(result)   



```

```python
#DAY 3

import numpy as np
hours_arr = np.array([2.5, 5.1, 3.2, 8.5, 3.5])
print(f"Studied hours: {hours_arr}")
print(type(hours_arr))

new_hours = hours_arr + 1.0
print(f"New hours: {new_hours}")

mean_hours = np.mean(hours_arr)
max_hours = np.max(hours_arr)
min_hours = np.min(hours_arr)
print(f"Average hours studied: {mean_hours}")
print(f"Highest hours studied: {max_hours}")
print(f"Lowest hours studied: {min_hours}")

############################################## 

scores_arr = np.array([21, 47, 27, 75, 30])
print(f"Scores: {scores_arr}")
mean_scores = np.mean(scores_arr)
min_scores = np.min(scores_arr)
print(f"Average score: {mean_scores}")
print(f"Minimum score: {min_scores}")

new_scores = scores_arr * 2
print(f"New scores: {new_scores}")
```
